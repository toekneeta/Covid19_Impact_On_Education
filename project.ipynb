# Research Question

 From the initial shift to online learning (March 2020) until today, what variables—teacher’s involvement (measured by absences), technological skills, financial circumstances, reliable internet access—has had an effect on the mental well-being of students (measured through student surveys)?

<a id='bg'></a>
## Background & Prior Work

An academic study from Plos One has shown that teacher engagement improves online learning outcomes for students. A teacher’s engagement includes their attitudes, giving timely feedback, answering questions, and delivering online lectures, which all leaves a student satisfied with their remote learning experience (Keržič et al, 2021) (1). This study introduces less influential factors to student well-being, such as student social media and technology skills. Students’ expertise in computer use, learning platforms, and social media all influences their participation and satisfaction in e-learning. The study uses the DeLone and McLean’s Model (D&M model) of Information Systems Success to show that the frequency of a student using online systems has a direct connection to improving their satisfaction and online academic performance. Both of these findings are relevant to our project since a teacher’s role and social media skills are both variables that we are also considering to analyse.


A study done in the Netherlands has shown that the circumstances surrounding students academic success at school is due to their immediate surroundings (ie. learning from a disadvantaged home) and not so much due to the things the schools or the district is doing to help. This shows that family income negatively affects academic performance (Engzell et al, 2021) (2). Lower income students who were unable to pay for their studies were more likely to report poor academic performance and worries about their family and relationships compared to high-income, part-time students. Although students reported high confidence with online learning activities, only 59.9 percent of students had access to a good internet connection in the global sample (Cameron et al 2021) (3). These findings merit the investigation of family income as a key variable in investigating the factors affecting student well being.

All of our variables that are of type `string` are considered to have order, and so it makes sense to assign them `int` values.

To tidy this column, we will create new columns, each representing a possible option, where the possible values for these columns are `1` if the participant selected that option or `0` if the participant did not select that option.

All variables that are currently of type `string` will be converted to type `int`: 
- `Purpose for using social media networks` will be broken up to have a column for each possible option, where `1` indicates that the participant selected that option and `0` indicates that the participant did not select that option
- `Internet availability`, `Internet Affordability`, `Power Availability`, and `Technical Skills` are measured by the likert scale: `Strongly Agree, Agree, Neutral, Disagree, Strongly Disagree`, will be converted to `1, 2, 3, 4, 5`, respectively

```python
# Initial look at the social_media dataset
social_media.head()
```

```python
# Selects only the columns that will be used
social_media = social_media[['Why do you use social networks? (You can choose more than one)', \
              'Internet availability and signal strength will be a problem in using the social media and emerging technologies for learning  ', \
              'Internet data bundles affordability will be a problem in using the social media and emerging technologies for learning  ', \
              'Power availability will be a problem in using the social media and emerging technologies for learning  ', \
              'I have the technical skills to use social media platforms and emerging technologies for learning']]

social_media.head()
```

```python
# Renames the columns to make them more readable
social_media.columns = ['purpose', 'internet_availability', 'internet_affordability', 'power_availability', 'tech_skill']
social_media.head()
```

As shown below, there are no missing values in our data, so we will not be dropping any observations.

```python
social_media.isnull().sum()
```

The variables that are of type `string` are considered to have order will be assigned to have `int` values.

The variables `internet_availability, internet_affordability, power_availability, tech_skill` are all measured by the likert scale, and so `Strongly Disagree, Disagree, Neutral, Agree, Strongly Agree`, will be converted to `1, 2, 3, 4, 5`, respectively. 

```python
# This function converts the likert scale to int values
def convert_scale(string):
    if string == 'Strongly Agree':
        return 5
    elif string == 'Agree':
        return 4
    elif string == 'Neutral':
        return 3
    elif string == 'Disagree':
        return 2
    return 1
```

```python
# Applying convert_scale function to internet_availability column
social_media = social_media.assign(internet_availability=social_media.get('internet_availability').apply(convert_scale))

# Applying convert_scale function to internet_affordability column
social_media = social_media.assign(internet_affordability=social_media.get('internet_affordability').apply(convert_scale))

# Applying convert_scale function to power_availability column
social_media = social_media.assign(power_availability=social_media.get('power_availability').apply(convert_scale))

# Applying convert_scale function to tech_skill column
social_media = social_media.assign(tech_skill=social_media.get('tech_skill').apply(convert_scale))

social_media
```

When the participants answered the question corresponding to the `purpose` column, they were allowed to select as many options as they wanted. 

We will separate each possible response within `purpose` into a column in the `social_media` DataFrame, where a `1` will be assigned if the participant selected that option and a `0` will be assigned if they didn't.

```python
# dictionary with the option as the key and the corresponding new column as the value
all_options = {'To find information': 'pur_find_info', 'To play games': 'pur_play_games', \
               'To make professional and business contacts': 'pur_make_contacts', \
               'To keep in touch with family and friends': 'pur_fam_friends', 'To make new friends': 'pur_make_friends', \
               'To get latest information/gist': 'pur_latest_info', 'To share videos/ pictures/ music': 'pur_share_media', \
               'To share your experience': 'pur_share_exp', 'For academic purposes': 'pur_academic', \
               'To while away time': 'pur_away_time'}

# creating a column for each possible option for the purpose column, with a default value of 0 for each column
social_media = social_media.assign(pur_find_info=0)
social_media = social_media.assign(pur_play_games=0)
social_media = social_media.assign(pur_make_contacts=0)
social_media = social_media.assign(pur_fam_friends=0)
social_media = social_media.assign(pur_make_friends=0)
social_media = social_media.assign(pur_latest_info=0)
social_media = social_media.assign(pur_share_media=0)
social_media = social_media.assign(pur_share_exp=0)
social_media = social_media.assign(pur_academic=0)
social_media = social_media.assign(pur_away_time=0)

# function that assigns 1 to an option column if the participant chose that column
def break_up_purpose(row):
    selected_options = row['purpose'].split(';')
    for i in all_options:
        if i in selected_options:
            row[all_options[i]] = 1
    return row

# applies the break_up_purpose function to social_media DataFrame and removes the purpose column afterwards
social_media = social_media.apply(break_up_purpose, axis=1)
social_media = social_media.drop('purpose', axis=1)
```

Only columns that were seen as important to our project were kept, and the values in the columns were converted from the likert scale to a numerical counterpart. ### Exploring Teacher Absences

The first variable we explored was teacher absences by seeing how often students reached out to professors, implying their availablity during remote learning. The x-axis range in this first section is scaled on how often a student talked to a professor, 1 being never and 6 being very often.

```python
sns.countplot(x='talked_prof', data=questionnaire)
plt.title('How Often Students Talked to Professors in Questionnaire Sample');
```

Because the graph is right skewed, this graph tells us that, in general, students did not talk to professors very often during the pandemic. To see why this is the case, we will continue to explore how often students talked to professors and try to see if there are any confounding variables.

Below, the variable `mental_health_worries` is scaled from 1 being the never, and 5 being always.

```python
sns.countplot(x='talked_prof', hue='mental_health_worries', data=questionnaire)
plt.title('How often Students Talked to Professors by Mental Health Worries');
```

This graph explores if mental health worries is a counfounding variable. Students who never talked to professors reported almost no mental health worries. Students who did talk to professors, also barely reported mental health worries. Therefore, there doesn't seem to be much correlation between how often students talked to their professors and their mental health worries. 

The second counfounding variable we explored in what impacts the act of reaching out to professors was worries related to studying. 

```python
sns.countplot(x='talked_prof', hue='studying_worries', data=questionnaire)
plt.title('How often Students Talked to Professors by Academic Worries');
```

This graph seemingly shows no trend between how often a student talked to their professor during the pandemic and how stressed out they were by their studies. It is interesting to see that students who were worried by their academics didn't talk to their professors more often. Worries of study appears to indepedent of talking to teachers.

```python
sns.countplot(x='talked_prof', hue='future_pandemic_worries', data=questionnaire)
plt.title('How often Students Talked to Professors by Pandemic Worries');
```

The above graph explores one last potenial confound: pandemic worries. However, it doesn't seem to correlate to how often students talked to a professor. Moreover, people who were less worried about the pandemic make up more of the people who did not talk to the professors. 

Next, we did a simple comparison betweeen students who talked to their professors and those who talked to their classmates, to see if it implied anything about teacher absenses. 

```python
sns.countplot(x='talked_prof', hue='talked_classmate', data=questionnaire)
plt.title('How often Students Talked to Professors vs. Classmates');
```

This graph reads that people who never talked to their professors also overwhelmingly never talked to their classmates. On the other hand, people who almost always talked to their professors always talked to their classmates. The people who generally do not reach out professors, also will not reach out thier classmates. Both relationships are independent of each other.

# Statistical Analysis

Now that we've explored our data to get a good understanding of the spread and shape of our data, we can perform some statistical tests to try and confirm our observations.

To answer our initial question, we will compare each of the variables we're interested in - teacher’s involvement, technology skills, financial circumstances, reliable internet access - with a student's self-reported mental well-being.

The following variables in our `questionnaire` dataset will be used for the comparison:
- teacher's involvement: `talked_prof`
- technology skills: `skill_software`
- financial circumstances: `finances_worries`
- reliable internet access: `internet`
- mental well-being: `mental_health_worries`

Considering that each of the variables above are all measured on the likert scale, they are all ordinal variables. This means that any statistical test we run must be non-parametric.

This means that we can run a **chi-square test of independence**, which is a non-parametric statistical test that allows us to compare 2 ordinal variables and see if there is a dependent relationship between the two variables.

To run a chi-square test of independence, we must first check that we meet the following assumptions of the test.

1. Each observation is independent of all the others.

   This assumption is met since each of the observations in our dataset are randomly sampled, where there is no connection between the survey responses of one participant to another.
   
   
2. Each observation cannot be in more than 1 group in any variable.

   This assumption is met since participants in the survey were only allowed to choose one of the options for each variable. Thus, no participants will have selected multiple groups for a variable.
   
   
3. All expected values in our data must be at least 5.

   This assumption will be checked during each test when the expected values are created. If any expected values are less than 5, the chi-square function defined below will return False and our statistical test will not proceed.
    
Since our assumptions are met, we can proceed with our statistical test. For each of our four tests, we will run them at a significance level of $\alpha$ = 0.01

<a id='chi_sq'></a>
### Setting Up Chi-Square Test of Independence

We will create a function that allows us to calculate the value of chi-square. The chi-square value is calculated as followed:


## <center>${\chi}^2=\sum \frac{(O - E)^2}{E}$

   <center>with $(r-1) \cdot (c-1)$ degrees of freedom 

where **O** is the observed value from the contingency table and **E** is the expected value, calculated by the row total multiplied by the column total, and dividing that product by the sample size. **r** and **c** represent the number of rows and columns, respectively, in our contingency table. Our function will return `False` if assumption 3 is not met.

```python
def chi_square(cont_table, df, rows, columns):
    chi_square_value = 0
    for i in columns:
        for j in rows:
            O = cont_table[i][j] # observed value
            E = cont_table[i]['All'] * cont_table['All'][j] / cont_table['All']['All'] # expected value
            if E < 5: # checks assumption 3 of a chi-square test of independence
                return False
            chi_square_value += (O - E)**2 / E # calculates chi-square and adds to chi_square_value variable

    return chi_square_value
```

Using our chi-square value, we can calculate the p-value of our hypothesis test using the chi-squared distribution with $(r-1) \cdot (c-1)$ degrees of freedom.

We create a function that passes in the chi-squared value, calculates the p-value, and makes a decision on our hypothesis test at the significance level of $\alpha$ = 0.01

```python
def get_p_value(chi_square, rows, columns, alpha=0.01):
    p_value = 1 - stats.chi2.cdf(chi_square, (len(rows)-1)*(len(columns)-1))
    if p_value <= alpha:
        conclusion = "At a significance level of " + str(alpha) + \
                     ", we reject the null hypothesis in favor of our alternative hypothesis."
    else: 
        conclusion = "At a significance level of " + str(alpha) + \
                     ", we keep the null hypothesis."
    print("The p-value of our chi-square test of independence is " + str(p_value) + ".\n\n" + conclusion)
```

<a id='ti_chi_sq'></a>
### Testing Relationship between Teacher's Involvement and Mental Health Worries

Our hypotheses for our test are as follows:
- H<sub>0</sub> = There **is not** a relationship between a teacher's involvement and a student's mental health worries. (They are independent)
- H<sub>1</sub> = There **is** a relationship between a teacher's involvement and a student's mental health worries. (They are dependent)

We create a contingency table that displays the joint responses to our two variables.

```python
teacher_cont_table = pd.crosstab(questionnaire.talked_prof, questionnaire.mental_health_worries, margins=True)
teacher_cont_table
```

We then calculate the chi-square value using our `chi_square` function and using that value to find the p-value and to come up with a conclusion to our hypothesis test using our `get_p_value` function.

```python
rows = np.sort(questionnaire['talked_prof'].unique())
columns = np.sort(questionnaire['mental_health_worries'].unique())

chi_square_val = chi_square(teacher_cont_table, questionnaire, rows, columns)

if chi_square_val:
    get_p_value(chi_square_val, rows, columns)
else:
    print('Assumption 3 of chi-square test of independence not met.')
```

With a p-value less than 0.01, our statistical test concluded that there **is a relationship** between a teacher's involvement and a student's mental health worries.

COVID-19 has greatly impacted many facets of society. In particular, the pandemic ushered in an unprecedented period of remote learning that has marked the education journeys of thousands of students around the world. This project sheds light on just a few of the challenges that students have had to face during the remote learning period while simultaneously dealing with COVID-19, and shows how these factors have taken a toll on students’ mental health.
